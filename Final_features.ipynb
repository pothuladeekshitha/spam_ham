```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
nltk.download('stopwords')
nltk.download('punkt')
stop_words = set(stopwords.words('english'))

df['STOP WORDS']='0'
df['NUMBER OF STOP WORDS']='0'
for i in range(0, len(df.index)):
    x=[]
    word_tokens = word_tokenize(df.loc[i].at['Message'])
    filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
    for w in word_tokens:
        if w in stop_words:
            x.append(w)
    df.at[i,"STOP WORDS"]= x
    df.at[i,"NUMBER OF STOP WORDS"]= len(x)
    
    
df
```

```python
def extract_entities(doc, key):
    x=[]
    nlp = spacy.load("en_core_web_sm")    
    for ent in doc.ents:
        if ent.label_== key :
            x.append(ent.text)
    
    return x
```

Parts of speeech extraction

```python
df['ADJECTIVES']='0'
df['Number of ADJECTIVES']='0'
df['INTERJECTION']='0'
df['Number of INTERJECTION']='0'
df['NOUNS']='0'
df['Number of NOUNS']='0'
df['Verbs']='0'
df['Number of Verbs']='0'
df['WHwords']='0'
df['Number of WHwords']='0'
df['Foreign word']='0'
df['Number of Foreign word']='0'


import nltk
nltk.download('averaged_perceptron_tagger')

is_adj = lambda pos: pos[:2] == 'JJ' or pos[:2] == 'JJS' or pos[:2] == 'JJR'
for i in range(0, len(df.index)):
    x=[]
    tokenized = nltk.word_tokenize(df.loc[i].at['Message'])
    x = [word for (word, pos) in nltk.pos_tag(tokenized) if is_adj(pos)] 
    df.at[i,"ADJECTIVES"]= x
    df.at[i,"Number of ADJECTIVES"]= len(x)
    
    


is_noun = lambda pos: pos[:2] == 'NN' or pos[:2] == 'NNS' or pos[:2] == 'NNP' or pos[:2] == 'NNPS'
for i in range(0, len(df.index)):
    x=[]
    tokenized = nltk.word_tokenize(df.loc[i].at['Message'])
    x = [word for (word, pos) in nltk.pos_tag(tokenized) if is_noun(pos)] 
    df.at[i,"NOUNS"]= x
    df.at[i,"Number of NOUNS"]= len(x)
        
    
    
is_interjection = lambda pos: pos[:2] == 'UH' 
for i in range(0, len(df.index)):
    x=[]
    tokenized = nltk.word_tokenize(df.loc[i].at['Message'])
    x = [word for (word, pos) in nltk.pos_tag(tokenized) if is_interjection(pos)] 
    df.at[i,"INTERJECTION"]= x
    df.at[i,"Number of INTERJECTION"]= len(x)
    
    
    
    
is_wh = lambda pos: pos[:2] == 'WDT' or pos[:2] == 'WP' or pos[:2] == 'WD$' or pos[:2] == 'WRB' 
for i in range(0, len(df.index)):
    x=[]
    tokenized = nltk.word_tokenize(df.loc[i].at['Message'])
    x = [word for (word, pos) in nltk.pos_tag(tokenized) if is_wh(pos)] 
    df.at[i,"WHwords"]= x
    df.at[i,"Number of WHwords"]= len(x)
    

is_fw = lambda pos: pos[:2] == 'FW' 
for i in range(0, len(df.index)):
    x=[]
    tokenized = nltk.word_tokenize(df.loc[i].at['Message'])
    x = [word for (word, pos) in nltk.pos_tag(tokenized) if is_wh(pos)] 
    df.at[i,"Foreign word"]= x
    df.at[i,"Number of Foreign word"]= len(x)


is_verb = lambda pos: pos[:2] == 'VB' or pos[:2] == 'VBD' or pos[:2] == 'VBG' or pos[:2] == 'VBZ' 
for i in range(0, len(df.index)):
    x=[]
    tokenized = nltk.word_tokenize(df.loc[i].at['Message'])
    x = [word for (word, pos) in nltk.pos_tag(tokenized) if is_verb(pos)] 
    df.at[i,"Verbs"]= x   
    df.at[i,"Number of Verbs"]= len(x)
    
```
