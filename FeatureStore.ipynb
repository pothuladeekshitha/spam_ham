```python
from feast import FeatureStore
from feast.infra.offline_stores.file_source import SavedDatasetFileStorage

store = FeatureStore(repo_path='.')

entity_df = pd.read_parquet(path = 'data/target_df.parquet')

training_data = store.get_historical_features(
entity_df = entity_df,
    features = [
        "predict_df_feature_view:Number of exclamations",
        "predict_df_feature_view:Question Marks",
        "predict_df_feature_view:NUMBER OF STOP WORDS",
        "predict_df_feature_view:number of dates",
        "predict_df_feature_view:number of Cardinals",
        "predict_df_feature_view:Number of ADJECTIVES",
        
        "predict_df_feature_view:Number of digits",
        "predict_df_feature_view:Number of NOUNS",
        "predict_df_feature_view:Number of Verbs",
        "predict_df_feature_view:Number of WHwords",
        "predict_df_feature_view:STOP WORDS",
        "predict_df_feature_view:ORG",
        "predict_df_feature_view:CARDINAL",
        "predict_df_feature_view:MONEY",
        "predict_df_feature_view:ADJECTIVES",
        "predict_df_feature_view:NOUNS",
        "predict_df_feature_view:WHwords",
        "predict_df_feature_view:Verbs",
        "predict_df_feature_view:DATE",
         
               ]
)

dataset = store.create_saved_dataset(
from_=training_data,
    name = "feature_storeData",
    storage = SavedDatasetFileStorage('data/feature_storeData.parquet')
) 
```
